```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sudarshan24byte/online-food-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(f"{path}/onlinefoods.csv")
df
```

```python
# 6. List all categorical and numerical features separately.


numerical_features = df.select_dtypes(include=['number']).columns.tolist()

categorical_features = df.select_dtypes(include=['object', 'category']).columns

print("Numerical Features:", numerical_features)
print("Categorical Features:", categorical_features)
```

```python
# 7. Plot the distribution of customer ages using a histogram

plt.hist(df["Age"], bins=20, color="blue",edgecolor = "black")
plt.title("Distribution of Customer Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# 8. Create a count plot showing the number of customers in each occupation.




sns.countplot(x = "Occupation", data = df ,palette= "Oranges")
plt.title("Number of Customers by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()


```

```python
#9. Plot the distribution of monthly income categories.

sns.countplot(x = "Monthly Income" , hue = "Monthly Income" , data = df ,palette = "Set3" , legend = True)

plt.title("Distribution of Monthly Income Categories")
plt.xlabel("Monthly Income")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

```

```python
 # 10. Find the top 5 most common educational qualifications.

df["Educational Qualifications"].value_counts()
```

```python
#11. Compare the distribution of Gender across different Occupations using an appropriate visualization

sns.countplot(x = "Occupation" , hue = "Gender" , data = df , legend = True)
plt.title("Gender Distribution Across Occupations")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

```

```python
# 12. Analyze the relationship between Marital Status and Feedback.

sns.countplot(x = "Marital Status" , hue = "Feedback" , data = df)
plt.title ("Relationship between mariatal Status and Feedback")
plt.xlabel
```

```python
# 13. Compare Family Size across different Monthly Income groups using a box plot.

sns.boxplot(data = df , x = "Monthly Income" , hue = "Monthly Income" , y = "Family size", palette ="Paired" )
plt.title("Compare Family Size accrss diffrent Monthly Income")
plt.xticks(rotation = 30)
plt.show()
```

```python
#14. Visualize the relationship between Age and Family Size using a scatter plot.

sns.scatterplot(x = "Age" , y = "Family size" , data=df ,color = "black")
plt.title("relationship between Age and Family Size")
plt.show()
```

```python
#15. Create a heatmap of correlations among all numerical features.

numerical_feature = df.select_dtypes(include = ["int64" , "float64"])
correlation_matrix = numerical_feature.corr()
sns.heatmap(correlation_matrix,annot = True , cmap = "rocket" , fmt =".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#16. Analyze how Occupation, Monthly Income, and Feedback are related.

sns.countplot(data=df, x="Occupation",hue="Feedback",palette="Set2")

plt.title("Occupation And Feedback")
plt.xticks(rotation=45)
plt.show()

```

```python
#17. Compare the average family size for each educational qualification.

df.groupby("Educational Qualifications")["Family size"].mean()
```

```python
#18. Plot customer locations using Latitude and Longitude on a scatter plot and identify any visible clusters.

sns.scatterplot(data = df , x = "latitude" , y = "longitude" , color = "black")
plt.title("Custumer location")
plt.show()

```
